def png_to_jpg(png_path):
    fill_color = None
    
    for filename in listdir(png_path):
        if filename.endswith('png'):
            img = Image.open(png_path+filename)
            
            if img.mode in ('RGBA', 'LA'):
                background = Image.new(img.mode[:-1], img.size, fill_color)
                background.paste(img, img.split()[-1])
                img = background
                
                jpg_name = filename[:-4]+'.jpg'
                img.save(jpg_name)

```python
import os, glob, numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import keras.backend.tensorflow_backend as K

import tensorflow as tf
#config = tf.ConfigProto(device_count = { 'GPU': 0 })
#config.gpu_options.allow_growth = True
config = tf.ConfigProto(allow_soft_placement=True)
config.gpu_options.allocator_type = 'BFC'
config.gpu_options.per_process_gpu_memory_fraction = 0.90
config.gpu_options.allow_growth = True

session = tf.Session(config=config)
#session = tf.Session()
X_train, X_test, y_train, y_test = np.load('./dog__.npy',allow_pickle=True)
print(X_train.shape)
print(X_train.shape[0])
```

```python

categories = ["beagle", "golddog", "well", "hoon", "husky"]
nb_classes = len(categories)

#일반화
X_train = X_train.astype(float) / 255
X_test = X_test.astype(float) / 255
```

```python
with K.tf_ops.device('/device:GPU:0'):
    model = Sequential()
    model.add(Conv2D(32, (3,3), padding="same", input_shape=X_train.shape[1:], activation='elu'))
    model.add(MaxPooling2D(pool_size=(1,1)))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(64, (3,3), padding="same", activation='elu'))
    model.add(MaxPooling2D(pool_size=(1,1)))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(128, (3,3), padding="same", activation='elu'))
    model.add(MaxPooling2D(pool_size=(1,1)))
    model.add(Dropout(0.25))
    
    model.add(Flatten())
    model.add(Dense(10, activation='elu')) #원래 256
    model.add(Dropout(0.5))
    model.add(Dense(nb_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model_dir = './model'
    
    if not os.path.exists(model_dir):
        os.mkdir(model_dir)
    
    model_path = model_dir + '/multi_img_classification.model'
    checkpoint = ModelCheckpoint(filepath=model_path , monitor='val_loss', verbose=1, save_best_only=True)
    early_stopping = EarlyStopping(monitor='val_loss', patience=6)
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=12, epochs=50, validation_data=(X_test, y_test), callbacks=[checkpoint, early_stopping])
```

```python
#checkpoint
early_stopping
```

```python
print("정확도 : %.4f" % (model.evaluate(X_test, y_test)[1]))
```

```python
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len, y_vloss, marker='.', c='red', label='val_set_loss')
plt.plot(x_len, y_loss, marker='.', c='blue', label='train_set_l
         oss')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()
plt.show()
```
