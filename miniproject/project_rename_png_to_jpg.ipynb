```python
# -*- coding: utf-8 -*-
import sys
from os import rename, listdir
from PIL import Image

# https://118k.tistory.com/303
# 파일명 바꾸기
def changeName(path, cName):
    i = 1
    for filename in listdir(path):
        print(path+filename, '=>', path+str(cName)+str(i)+'.png')
        rename(path+filename, path+str(cName)+str(i)+'.png')
        i += 1
# 실행
# changeName('./husky/', 'husky')

# https://gist.github.com/abhbtbb/1cd050f001980ba5b306b293a7b2b0ca
# png 파일을 jpg로 바꾸기 (폴더 지정은 따로 못하는 것 같음ㅠ 현재 폴더로 저장됨)
def png_to_jpg(png_path):
    fill_color = None
    
    for filename in listdir(png_path):
        if filename.endswith('png'):
            img = Image.open(png_path+filename)
            
            if img.mode in ('RGBA', 'LA'):
                background = Image.new(img.mode[:-1], img.size, fill_color)
                background.paste(img, img.split()[-1])
                img = background
                
                jpg_name = filename[:-4]+'.jpg'
                img.save(jpg_name)
# 실행
# png_to_jpg('./husky/')
```
