```python
import pandas as pd
import numpy as np

data1 = pd.read_csv('대출.csv', index_col=0)
data2 = pd.read_csv('졸업자수.csv', index_col=0)
data3 = pd.read_csv('사교육비.csv', index_col=0)
data4 = pd.read_csv('인구.csv', index_col=0)
data5 = pd.read_csv('자가용.csv', index_col=0)
data6 = pd.read_csv('출산진료비.csv', index_col=0)
```

```python
data6.rename(index={'2016년':2016, '2015년':2015, '2014년':2014, '2013년':2013, '2012년':2012, '2011년':2011, '2010년':2010, '2009년':2009}, inplace=True)
```

```python
frames = [data1,data2, data3,data4,data5,data6]
```

```python
result = pd.concat(frames, axis=1, sort=False)
```

```python
data = pd.DataFrame(result)
data
```

```python
data.to_csv('data_merge1.csv', index=True)
```
