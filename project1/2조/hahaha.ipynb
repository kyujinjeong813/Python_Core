```python
from sklearn.feature_selection import RFE  # recursive feature elimination (재귀적 특징제거 )
from sklearn.svm import SVR

estimator = SVR(kernel = 'linear') #svc로 만든ㄹ었지만 선형과 같다  #Regression 예측  SVC : Classification
selector =RFE(estimator,16,step=1) #5개 변수만 남기겠다_ step : 한번에 하나씩 제거 ,5개를 없에야 하는 상황
#변수를 제거하는 이유 : 잡음제거 _s 영향력이 없는 변수들 
selector = selector.fit(new_data,target)
print(selector.support_,'\n')
print(selector.ranking_)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

data = pd.read_csv('team2.csv', index_col=0)
data
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
X,y
```

```python
model = StandardScaler().fit_transform(X)
print(model)
```

```python
features = model.T
covariance_matrix=np.cov(features)
print(covariance_matrix)
```

```python
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
```

```python
print('Eigenvectors \n',eig_vecs)
```

```python
print('Egenvalues \n',eig_vals)
```

```python
eig_vals[0]/sum(eig_vals)
```

```python
projected_X = model.dot(eig_vecs.T[0])
```

```python
projected_X
```

```python
result = pd.DataFrame(projected_X, columns=['PC1'])
result['y-axis']=0.0
result['label'] = y
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


data = pd.read_csv('team2.csv', index_col=0)
data
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
X,y
```

```python
# MinMax_Scale
from sklearn.preprocessing import minmax_scale
minmax_X = minmax_scale(X)
minmax_Y = minmax_scale(y)

X_train, X_test, y_train, y_test = train_test_split(minmax_X,minmax_Y, random_state=0)
print(minmax_X.shape,minmax_Y.shape,'\n')
print(minmax_X[:3,:],'\n')
print(minmax_Y)
```

```python
#모델만들기 2. SVR




model = SVR(gamma='scale')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)


# 모델평가
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("MSE",mean_squared_error(y_test, y_pred))
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))

df1=pd.DataFrame({'y_data':y_test,'preds':y_pred})
df1.corr()

# 과소인지 과대인지 확인해보자 ( )
print()
print(model)
X_pred=model.predict(X_train)
print('MSE', mean_squared_error(y_train, X_pred))
y_pred = model.predict(X_test)
print("MSE",mean_squared_error(y_test, y_pred))  # mse를 비교했을때 test>train : 과소적합 
```
