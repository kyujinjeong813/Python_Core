```python
real_df=pd.read_csv("real.csv",encoding='utf-8',index_col=0)
```

```python
real_df.count()
```

```python
del real_df['주택소유 신혼부부 수']
```

```python
del real_df['대출일반']
```

```python
del real_df['대출주택']
```

```python
real_df.count()  # 원래 결측치 갯수 
```

```python
real_df['1인가구 비율']=real_df['1인가구 비율'].fillna(method='ffill')  # 1인가구비율
```

```python
del real_df['출산율']
```

```python
baby= pd.read_csv("출생아수.csv",encoding='utf-8',index_col=0)
```

```python
rdf=pd.merge(baby, real_df , how='outer',left_index=True, right_index=True)
```

```python
del rdf['출생아수(명)']
```

```python
# 결측치 처리
```

```python
rdf['졸업자수']=rdf['졸업자수'].fillna(method='bfill')  # 졸업자수 
```

```python
rdf=rdf.fillna(method='bfill') # 모든 결측치 처리
```

```python
rdf=rdf.fillna(method='ffill')
```
